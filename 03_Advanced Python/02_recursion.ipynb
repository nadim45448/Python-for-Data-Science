### Multiplication of two number

```python
# multiplication of two number
# 4*5 =20 = 4+4+4+4+4
# 4+(4*4)
# 4+4+(4*3)
# 4+4+4+(4*2)
# 4+4+4+4+(4*1)
# a+a*(b-1)

# iterative
def multiply(a, b):
  result = 0
  for i in range(b):
    result = result + a
  print(result)

multiply(4,5)
```

```python
# recursive
def multiply(a,b):
  if b==1:
    return a
  else:
    return a+multiply(a, b-1)

print(multiply(4,5))
```

### Factorial of a given number

```python
# 0! = 1
# 1! = 1
# 5! = 5*4

def factorial(num):
  if num == 0 or num == 1:
    return 1
  else:
    return num * factorial(num - 1)

print(factorial(5))
```

### Palindrome

```python
def palindrome(st):
  if len(st)<=1:
    print("Palindrome")
  else:
    if st[0]==st[-1]:
      palindrome(st[1:-1])
    else:
      print("Not palindrome")

palindrome("madam")
palindrome("python")
palindrome("abba")
```

### Fibonacci

```python
def fib(n):
  if n<=1:
    return 1
  else:
    last = fib(n-1)
    first = fib(n-2)
    return first + last

print(fib(5))
```
