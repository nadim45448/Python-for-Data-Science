# How objects access attributes

```python
class Person:
  def __init__(self, name,country):
    self.name = name
    self.country = country

  def greet(self):
    if self.country == 'Bangladesh':
      print(f'Assalamu alaikum {self.name}')
    else:
      print(f'welcome {self.name}')

```

```python
# how to access attributes
p = Person('nadim','Bangladesh')
```

```python
p.name
```

```python
# how to access methods
p.greet()
```

```python
# what if i try to access non-existent attributes
p.gender
```

# Attribute creation from outside of the class

```python
p.gender
```

### Reference Variables

- Reference variables hold the objects
- We can create objects without reference variable as well
- An object can have multiple reference variables
- Assigning a new reference variable to an existing object does not create a new object

```python
# object without a reference
class Person:

  def __init__(self):
    self.name = 'nadim'
    self.gender = 'male'

```

```python
p = Person()
# p is a reference variable that refers to the object created in memory.
# The actual object (instance of Person) is stored in memory, and p holds its reference (address).
# So, the object exists in memory, and p is just a name (reference) pointing to it.
```

```python
# object without a reference
Person()
```

```python
# Multiple ref
p = Person()
q = p

print(id(p))
print(id(q))
```

```python
print(p.name)
print(q.name)
```

```python
# change attribute value with the help of 2nd object
q.name = 'hossain'
print(q.name)
print(p.name)
```

# Pass by reference

```python
class Person:

  def __init__(self,name,gender):
    self.name = name
    self.gender = gender

# outside the class-> function
def greet(person):
  print('Hi my name is',person.name,'and I am a',person.gender)
  p1 = Person('hossain', 'male')
  return p1
```

```python
p = Person('nadim','male')
# greet(p)
x = greet(p)
print(x.name)
print(x.gender)
```

```python
class Person:

  def __init__(self,name,gender):
    self.name = name
    self.gender = gender

# outside the class -> function
def greet(person):
  print(id(person))
  person.name = 'hossain'
  print(person.name)

p = Person('nadim','male')
print(id(p))
greet(p)
print(p.name)
```

### Object ki mutability

```python
class Person:

  def __init__(self,name,gender):
    self.name = name
    self.gender = gender

# outside the class -> function
def greet(person):
  person.name = 'hossain'
  return person

p = Person('nadim','male')
print(id(p))
p1 = greet(p)
print(id(p1))
```
